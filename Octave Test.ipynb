```python
import numpy as np
import pandas as pd
```

```python
janFlight = pd.read_csv('JanFlight.csv')
febFlight = pd.read_csv('FebFlight.csv')
marFlight = pd.read_csv('MarFlight.csv')
aprFlight = pd.read_csv('AprFlight.csv')
mayFlight = pd.read_csv('MayFlight.csv')
junFlight = pd.read_csv('JunFlight.csv')
```

```python
janFlight.head(5)
```

```python
flights = pd.concat([janFlight, febFlight, marFlight, aprFlight, mayFlight, junFlight])
```

```python
flights = flights.fillna(value = 0)
#flights
```

```python
flights.columns
```

```python
firstQuarter = flights[flights['QUARTER'] == 1]
secondQuarter = flights[flights['QUARTER'] == 2]
```

```python
#Question 1:
#What airline had the most number of flights in the first quarter of 2019 ? (5 Points) *
#The airline that had most number of travels between January - March 2019

ops = firstQuarter.groupby('OP_UNIQUE_CARRIER').sum()
ops[ops['FLIGHTS'] == ops['FLIGHTS'].max()]['FLIGHTS']
```

```python
#Question 2:
#What is the total number of Airline Travels in the first half of 2019 ? (5 Points) *
#The total number of flights between January 2019 and June 2019
flights['FLIGHTS'].sum()
```

```python
#Question 3
#What is the Total AirTime by all Airlines in May 2019 ? (5 Points) *
#The result should be presented in hours

print(round(mayFlight['AIR_TIME'].sum()/60,2), 'hours')
```

```python
#Question 4
#What city do you consider as the busiest city in April 2019 ? (5 Points) *
#Mention the city name in CAPITAL LETTERS

busyCity = aprFlight.groupby('ORIGIN_CITY_NAME').sum()
busyCity[busyCity['FLIGHTS'] == busyCity['FLIGHTS'].max()]['FLIGHTS']
```

```python
#Question 5
#What airline covered the most distance in June 2019 ? (5 Points) *
#Mention the Airline Name (not code) in CAPITAL LETTERS

distCov = junFlight[['OP_UNIQUE_CARRIER', 'DISTANCE']].groupby('OP_UNIQUE_CARRIER').sum()
distCov[distCov['DISTANCE'] == distCov['DISTANCE'].max()]
```

```python
#Question 6
#What is the Tail Number of the aircraft that covered the most distance in Q2 of 2019 ? (5 Points) *
#Mention the Tail Number of the aircraft with most distance (miles) between Apr - Jun 2019
#in CAPITAL LETTERS

mostDist = secondQuarter[secondQuarter['OP_UNIQUE_CARRIER'] == 'WN']
mostDist

tailDist = mostDist[['TAIL_NUM', 'DISTANCE']].groupby('TAIL_NUM').sum()
tailDist[tailDist['DISTANCE'] == tailDist['DISTANCE'].max()]
```

```python
#Question 7
#Which state recorded the most delays in the 2nd Quarter of 2019 (5 Points) *
#Mention the name of the state in CAPITAL LETTERS

stateDelay = secondQuarter.groupby('ORIGIN_STATE_NM').sum()

stateDelay['Total'] = stateDelay['DEP_DELAY_NEW'] + stateDelay['ARR_DELAY_NEW']
stateDelay[stateDelay['Total'] == stateDelay['Total'].max()]['Total']
```

```python
#Question 8
#What is the daily average Taxi-out time in New York in February 2019 ? (5 Points) *
#Enter the average daily taxi-out time for New York

febFlight[febFlight['DEST_STATE_NM'] == 'New York']['TAXI_OUT'].mean()
```

```python
#Question 9
#How many airlines with New York destinations were diverted in Q1 2019 ? (5 Points) *
#No of New York bound airlines that were diverted

firstQuarter[firstQuarter['DEST_CITY_NAME'] == 'New York, NY']['DIVERTED'].sum()
```

```python
#Question 10
#Which of these delays is more rampant in Washington in the first half of 2019(5 Marks) *
washington = flights[flights['ORIGIN_CITY_NAME'] == 'Washington, DC']
washington[['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY','SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']].sum()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import cufflinks as cf

from plotly.offline import download_plotlyjs,init_notebook_mode, iplot, plot
init_notebook_mode(connected=True)
cf.go_offline()
#'OP_UNIQUE_CARRIER', 'TAXI_OUT', 'ORIGIN_CITY_NAME', 'ORIGIN_STATE_NM'
```

```python
#Please provide analysis and charts on the distributions of Taxi-out time across states, 
#cites, and on the airlines as it affects fuel consumption. 
#Use data from the first half of 2019 only *
#TAXI OUT TIME OF AIRLINE OPERATORS
decision = flights[['OP_UNIQUE_CARRIER', 'TAXI_OUT', 'ORIGIN_CITY_NAME', 'ORIGIN_STATE_NM']]
decision
DecsOps = decision.groupby('OP_UNIQUE_CARRIER').sum()
DecsOps.iplot(y='TAXI_OUT', kind = 'bar', title = 'TAXI OUT TIME VERSUS AIRLINE OPERATORS (JAN - JUN)', 
              xTitle = 'Airlines Operators', yTitle= 'Taxi-Out Time (Mins)')
```

```python
#TAXI OUT TIME IN EACH CITY
DecsCity = decision.groupby('ORIGIN_CITY_NAME').sum()
DecsCity.iplot(y='TAXI_OUT', kind = 'bar', title = 'TAXI OUT TIME ACROSS CITIES (JAN - JUN)', 
               xTitle = 'Cities', yTitle= 'Taxi-Out Time (Mins)')
```

```python
#TAXI OUT TIME ACROSS STATES
DecsStates = decision.groupby('ORIGIN_STATE_NM').sum()
DecsStates.iplot(y='TAXI_OUT', kind = 'bar', title = 'TAXI OUT TIME ACROSS STATES (JAN - JUN)', 
                 xTitle = 'States', yTitle= 'Taxi-Out Time (Mins)')
```
